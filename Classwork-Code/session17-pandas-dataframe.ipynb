```python
# toss winner is match winner in percentage 
# ipl.head(1) 
# my code : using value_counts 
toss = (ipl['TossWinner'] == ipl['WinningTeam']).value_counts()

result = (toss[0] / toss.sum()) * 100
print(round(result,2))
```

```python
# write a Function that can return the track record of 2 teams against each other 
def track_record(team1,team2):
    record = ipl[(ipl['Team1'] == team1) & (ipl['Team2'] == team2)]
    team1_winner = record[(record['WinningTeam'] == team1)].shape[0]
    team2_winner = record[(record['WinningTeam'] == team2)].shape[0]
    return team1_winner, team2_winner
```

```python
team1 = input('Enter the First Team: ')  # Chennai Super Kings	# Royal Challengers Bangalore
team2 = input('Enter the Second Team: ')  # Mumbai Indians  # Kolkata Knight Riders
track_record(team1,team2)
```
